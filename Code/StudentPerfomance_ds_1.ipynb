```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
import pandas as pd

# DATA PREPROCESSING
# Label encoding for ordinal variables
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to each column
columns_to_encode = ['Access_to_Resources', 'Extracurricular_Activities', 'Sleep_Hours',
                     'Previous_Scores', 'Motivation_Level', 'Internet_Access', 'Tutoring_Sessions',
                     'Family_Income', 'Teacher_Quality', 'School_Type_1', 'Peer_Influence',
                     'Physical_Activity', 'Learning_Disabilities', 'Parental_Education_Level',
                     'Distance_from_Home', 'Gender_1']

for col in columns_to_encode:
    df[col] = le.fit_transform(df[col])

print(df.head())
```

```python

from sklearn.svm import SVC, SVR
from sklearn.model_selection import train_test_split, cross_val_score, KFold, StratifiedKFold
from sklearn.metrics import accuracy_score, mean_squared_error, confusion_matrix
import statsmodels.api as sm
import scipy.stats as stats

# ------------------------ SVM Classification ------------------------

def svm_classification(X, y_class):
    # Initialize the SVM model for classification (Support Vector Classification)
    model_class = SVC(kernel='rbf')

    # Set up 10-fold cross-validation for classification (StratifiedKFold for preserving class distribution)
    kfold_class = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    # Evaluate the model for classification using cross-validation
    cv_results_class = cross_val_score(model_class, X, y_class, cv=kfold_class, scoring='accuracy')

    # Print cross-validation results for classification
    print(f"Accuracy per fold for classification: {cv_results_class}")
    print(f"Mean Accuracy for classification: {np.mean(cv_results_class):.3f}")
    print(f"Standard Deviation for classification: {np.std(cv_results_class):.3f}")

    # Visualizing the results for classification (accuracy scores)
    plt.figure(figsize=(8, 6))
    plt.boxplot(cv_results_class, vert=False)
    plt.title('K-Fold Cross-Validation Accuracy for SVM Classification')
    plt.xlabel('Accuracy')
    plt.show()

    # Split the data for testing the classification model
    X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

    # Train the classification model
    model_class.fit(X_train, y_train)
    y_pred_svc = model_class.predict(X_test)

    # Evaluate the classification model
    accuracy = accuracy_score(y_test, y_pred_svc)
    print("\nSVM Classification (SVC) Performance:")
    print(f'Accuracy: {accuracy:.3f}')

    # Visualizing the confusion matrix for classification
    conf_matrix = confusion_matrix(y_test, y_pred_svc)
    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Low', 'Medium', 'High'], yticklabels=['Low', 'Medium', 'High'])
    plt.title('Confusion Matrix: SVM Classification (SVC)')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# ------------------------ SVM Regression ------------------------

def svm_regression(X, y_reg):
    # Initialize the SVR model (Support Vector Regression)
    model_reg = SVR(kernel='rbf')

    # Set up 10-fold cross-validation for regression
    kfold_reg = KFold(n_splits=10, shuffle=True, random_state=42)

    # Evaluate the model for regression using cross-validation (MSE)
    cv_results_reg = cross_val_score(model_reg, X, y_reg, cv=kfold_reg, scoring='neg_mean_squared_error')

    # Print cross-validation results for regression
    print(f"Mean Squared Error per fold for regression: {-cv_results_reg}")
    print(f"Mean MSE for regression: {-np.mean(cv_results_reg):.3f}")
    print(f"Standard Deviation for regression: {np.std(cv_results_reg):.3f}")

    # Visualizing the results for regression (MSE scores)
    plt.figure(figsize=(8, 6))
    plt.boxplot(-cv_results_reg, vert=False)
    plt.title('K-Fold Cross-Validation MSE for SVM Regression')
    plt.xlabel('Mean Squared Error')
    plt.show()

    # Fit the SVR model (Support Vector Regression)
    model_reg.fit(X_train, y_train)

    # Predict the values
    y_pred_reg = model_reg.predict(X_test)

    # Calculate residuals
    residuals_svr = y_test - y_pred_reg

    # Residual diagnostics for regression
    # 1. Linearity Check (Predicted vs Residuals plot)
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred_reg, residuals_svr, color='blue')
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Linearity Check: Predicted vs Residuals (SVR)')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    plt.show()

    # 2. Homoscedasticity Check (Residuals vs Fitted Values plot)
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred_reg, residuals_svr, color='blue')
    plt.axhline(0, color='red', linestyle='--')
    plt.title('Homoscedasticity Check: Residuals vs Fitted (SVR)')
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.show()

    # 3. Autocorrelation Check using Durbin-Watson test
    dw_statistic_svr = sm.stats.durbin_watson(residuals_svr)
    print(f"Durbin-Watson statistic for autocorrelation (SVR): {dw_statistic_svr:.3f}")

    # 4. Normality Check using Q-Q plot
    plt.figure(figsize=(8, 6))
    stats.probplot(residuals_svr, dist="norm", plot=plt)
    plt.title('Normality Check: Q-Q Plot (SVR)')
    plt.show()

    # 5. Normality Check using Histogram
    plt.figure(figsize=(8, 6))
    plt.hist(residuals_svr, bins=20, color='blue', edgecolor='black')
    plt.title('Normality Check: Residuals Histogram (SVR)')
    plt.xlabel('Residuals')
    plt.ylabel('Frequency')
    plt.show()

# ------------------------ Main Execution ------------------------

# Call the functions
svm_classification(X, y_class)  # Classification Task
svm_regression(X, y_reg)        # Regression Task
```
